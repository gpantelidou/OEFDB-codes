```python
import pandas as pd
```

```python
df = pd.read_excel("SupplyChainEmissionFactorsforUSIndustriesCommodities v1.1.xlsx", sheet_name="2016_Detail_Commodity")
```

```python
df
```

```python
df_carbon = df[df["Substance"] == "Carbon dioxide"]
df_methane = df[df["Substance"] == "Methane"]
df_no = df[df["Substance"] == "Nitrous oxide"]
df_other = df[df["Substance"] == "other GHGs"]
```

```python
display(df_carbon)
```

```python
GHG_df = pd.DataFrame(columns=["Commodity  Name", "Carbon Dioxide", "Methane", "Nitrous Oxide", "other GHGs"])
```

```python
GHG_df
```

```python
GHG_df = pd.DataFrame(columns=["Commodity Code", "Carbon Dioxide", "Methane", "Nitrous Oxide", "other GHGs"])
for i in range(len(df_carbon)):
    temp_values = {"Commodity Code" : df_carbon["Commodity Name"].iloc[i],
                   "Carbon Dioxide" : df_carbon["Supply Chain Emission Factors with Margins"].iloc[i],
                   "Methane" : df_methane['Supply Chain Emission Factors with Margins'].iloc[i],
                   "Nitrous Oxide" : df_no['Supply Chain Emission Factors with Margins'].iloc[i],
                   "other GHGs" : df_other['Supply Chain Emission Factors with Margins'].iloc[i]}
    GHG_df = GHG_df.append(temp_values, ignore_index=True)
```
